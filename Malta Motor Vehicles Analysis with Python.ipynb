### HTTP Request Result

The NSO report page returned status code 403 when accessed with Python requests.

This means the website blocks direct automated access to the page.

For this project, the data will be collected from the publicly available NSO report content and entered into a small structured dataset manually.

This keeps the project beginner-friendly and aligned with the HTTP learning


```python
motor_vehicle_data = [
    {
        "period": "Q4/2025",
        "licensed_motor_vehicles": 457403,
        "increase_from_previous_quarter": 3265
    },
    {
        "period": "Q1/2026",
        "licensed_motor_vehicles": 460648,
        "increase_from_previous_quarter": 3245
    }
]

print(motor_vehicle_data)
```

```python
import pandas as pd

df = pd.DataFrame(motor_vehicle_data)

df
```

```python
df.info()
```

```python
df.describe()
```

### Observation

Licensed motor vehicles increased from 457,403 in Q4/2025 to 460,648 in Q1/2026. The highest value was recorded in Q1/2026, while the lowest was recorded in Q4/2025, showing a positive upward trend in vehicle registrations during the period.

```python
growth = (
    (df.loc[1, "licensed_motor_vehicles"] -
     df.loc[0, "licensed_motor_vehicles"])
    / df.loc[0, "licensed_motor_vehicles"]
) * 100

print(f"Growth from Q4/2025 to Q1/2026: {growth:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.plot(
    df["period"],
    df["licensed_motor_vehicles"],
    marker="o"
)

plt.title("Licensed Motor Vehicles in Malta")
plt.xlabel("Quarter")
plt.ylabel("Number of Vehicles")

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.plot(
    df["period"],
    df["licensed_motor_vehicles"],
    marker="o"
)

for x, y in zip(df["period"], df["licensed_motor_vehicles"]):
    plt.text(x, y, f"{y:,}", ha="center")

plt.title("Licensed Motor Vehicles in Malta")
plt.xlabel("Quarter")
plt.ylabel("Number of Vehicles")

plt.grid(True)

plt.savefig("licensed_motor_vehicles_line_chart.png", bbox_inches="tight")

plt.show()
```

### Chart Observation

The number of licensed motor vehicles increased from 457,403 in Q4/2025 to 460,648 in Q1/2026. This represents an increase of 3,245 vehicles over the quarter, indicating continued growth in Malta's registered vehicle fleet.
